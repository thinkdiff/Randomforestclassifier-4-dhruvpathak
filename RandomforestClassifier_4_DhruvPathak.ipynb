```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
```

```python
# Load the dataset and inspect the first 5 rows
df = pd.read_csv('/content/breast-cancer.csv')
df.head()
```

### Data Cleaning

```python
df.info()
```

```python
# Drop the id column as it is not relevant
df.drop(['id'], axis=1, inplace=True)
df.head()
```

```python
# Convert the Diagnosis column to a numeric value, M = 1 and B = 0
df['diagnosis'] = df['diagnosis'].map({'M': 1, 'B': 0})
df.head()
```

### Feature selection

```python
# Separate the features and label
X = df.drop('diagnosis', axis=1)
y = df['diagnosis']
```

```python
# Scale the features to standardise the data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

### Split the data into training and test sets

```python
# Use 20% of the data for testing
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)
```

### Train using the RandomForest model

```python
model = RandomForestClassifier(n_estimators=100, random_state=42)
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
```

### Evaluate the model

```python
print("Accuracy:", accuracy_score(y_test, y_pred))
```

```python
print(classification_report(y_test, y_pred))
```

```python
print(confusion_matrix(y_test, y_pred))
```
